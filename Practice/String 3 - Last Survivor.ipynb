```python
n = 100
```

```python
nList = [i for i in range(1, n+1)]
start = nList[0]
removed = False
index = 0
```

**Brute Force Method**

```python
while len(nList) > 1:
    for i in range(0, len(nList)):
        if removed:
            nList[i] = 0
            removed = False
        else:
            removed = True
    
    nList = [i for i in nList if i != 0]
```
