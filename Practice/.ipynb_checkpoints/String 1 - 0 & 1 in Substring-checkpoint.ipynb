```python
A = "011100"

# "010111" --> 2
# "101000" --> 2
# "110100" --> 0
```

```python
def Solution(A):
  zeros = 0
  ones = 0
  counter = 0

  for a in A:
    if a == '0': zeros += 1
    else: ones += 1

    if zeros == ones:
      counter += 1
      zeros = ones = 0

  print(counter)
```

```python
def Solution(A):
  start = 0
  counter = 0

  for index in range(1, len(A)+1):
    if A[start:index].count('0') == A[start:index].count('1'): 
      counter += 1
      start = index + 1

  print(counter)
```

```python
Solution("1101001100")
```
